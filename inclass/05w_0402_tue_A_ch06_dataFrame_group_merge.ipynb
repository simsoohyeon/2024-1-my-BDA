```python
import pandas as pd

exam = pd.read_csv("data/exam.csv")
exam.sample(frac=.2)
```

```python
exam.agg('mean') ## 평균 알기
```

```python
exam.agg(mean_math = ('math', 'mean')) # 시험 예상
```

```python
import numpy as np
df = pd.DataFrame([[1, 2, 3],
                   [4, 5, 6],
                   [7, 8, 9],
                   [np.nan, np.nan, 10]],
                   columns=['A', 'B', 'C'])
df
```

```python
df.agg(['sum','mean','median','min','max','count'], axis=0)
```

```python
df.agg(['sum','mean','median','min','max','count'], axis=1) # 시험예상
```

```python
df.agg({'A' : ['sum', 'min'], 'B' : ['min', 'max']})
```

```python
df.agg('mean',axis=0)
```

```python
df.agg('mean',axis=1)
```

```python
df.agg('mean',axis="columns")
```

```python
df.agg(['sum','mean'], axis='columns')
```

```python
df.agg(['sum','mean'], axis='index')
```
