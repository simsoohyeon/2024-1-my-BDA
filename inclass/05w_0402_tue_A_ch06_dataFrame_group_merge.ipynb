#**집단별로 다시 집단 나누기**

```python
import pandas as pd

mpg = pd.read_csv('data/mpg.csv')
mpg.info()
```

```python
# 제조 회사(manufacturer) 및 구동 방식별(drv) 분리
# 도심연비(cty) 평균 구하기
mpg.groupby(["manufacturer", "drv"]).mean()[["cty"]]
```

```python
mpg.groupby(["manufacturer", "drv"])[["cty"]].mean()
```

```python
mpg.groupby(["manufacturer", "drv"]).agg(mean_cty = ('cty', 'mean'))
```

```python
# 추가: 제조회사, 구동방식별 분리, 도심연비 + 고속도로(hwy) 평균 연비
mpg.groupby(["manufacturer", "drv"]).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean'))
```

```python
mpg.groupby(["manufacturer", "drv"])[["cty"]].mean()
```

```python
mpg.groupby(["manufacturer", "drv"]).agg(mean_cty = ('cty', 'mean'))
```

```python
mpg
```

```python
## audi의 drv별 빈도

# audi 추출
# drv별 분리
# 빈도 구하기
mpg.query('manufacturer == "audi"').groupby('drv').agg( count = ('drv', 'count'))
```

```python
## chevrolet의 drv별 빈도

# chevrolet 추출
# drv별 분리
# 빈도 구하기
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).agg(n = ('drv', 'count'))
```

#**알아두면 좋은 value_count()로 집단별 빈도 간단하게 구하기**

```python
# mpg.groupby('drv'): 'mpg'
# DataFrame을 'drv' 열을 기준으로 그룹화합니다.
# agg(n = ('drv', 'count')):
# 각 그룹에 대해 'drv' 열의 값의 개수를 세는 집계 함수를 적용합니다.
# 이때, 새로운 열의 이름을 'n'으로 지정하고, 집계 함수로는 'count'를 사용합니다.

mpg.groupby('drv').agg(n = ('drv', 'count'))
```

#**pandas 함수 조합하기**

```python
# suv 추출
# 합산 연비 변수 만들기
# 제조 회사별로 분리
# 합산 연비 평균 구하기
# 내림차순 정렬
# 1~5위까지 출력, 시험 예상
mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
   .groupby('manufacturer') \
   .agg(mean_tot = ('total', 'mean')) \
   .sort_values('mean_tot', ascending = False)
```

```python
# suv 추출
# 합산 연비 변수 만들기
# 제조 회사별로 분리
# 합산 연비 평균 구하기
# 내림차순 정렬
# 1~5위까지 출력, 시험 예상
mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
   .groupby('manufacturer') \
   .agg(mean_tot = ('total', 'mean')) \
   .sort_values('mean_tot', ascending = False) \
   .head()
```

#**6-7 데이터 합치기**

```python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5],
                      'midterm' : [60, 80, 70, 90, 85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'    : [1, 2, 3, 4, 5],
                      'final' : [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
pd.merge(test1, test2)
```

```python
pd.merge(test1, test2, how = 'inner', on = 'id')
# id 열을 기준으로 조인, inner 내부조인
```
