## mpg 통합 연비 변수 만들기

```python
# mpg 데이터 불러오기
mpg = pd.read_csv("mpg.csv")
mpg.head()
```

#**시험 예상**

```python
# 통합 연비 변수(total) 만들기, total <= city + hmy의 평균
mpg['total'] = (mpg['cty']+mpg['hwy']) / 2
mpg.head()
```

```python
# 통합 연비 변수 만들기
mpg['mpg_mean'] = mpg[['cty','hwy']].mean(axis='columns')
mpg.head()
```

```python
sum(mpg['total']) / len(mpg) # total 합계를 행수로 나누기
```

```python
mpg['total'].mean() # 통합 연비 변수 평균
```

## 중첩 조건문 확인하기

```python
# total 기준으로 A,B,C 등급 부여
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
               np.where(mpg['total'] >= 20, 'B', 'C'))
# 데이터 확인
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rat=0)
```

```python
sns.countplot(mpg, x="grade", hue="grade",stat='percent', order=['A','B','C']);
```

```python
mpg['grade'].value_counts().sort_index().index
```

```python
sns.countplot(mpg, x="grade", hue="grade", order=mpg['grade'].value_counts().sort_index().index)
```
